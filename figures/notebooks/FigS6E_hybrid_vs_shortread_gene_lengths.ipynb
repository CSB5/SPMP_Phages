```python
import pandas as pd

df_hy_g = pd.read_csv('../data/Supplementary_Data_4_gene_annotations_of_SPMP_and_reference_vOTUs.tsv', sep='\t', usecols=['gene','length'])
df_hy_g = df_hy_g[df_hy_g.gene.str.startswith('TLL')] # only the SPMP vOTUs start with 'TLL'

df_sr_g = pd.read_csv('../data/hybrid_vs_shortread_comparison/shortread_votu_c50_genes.tsv', sep='\t', usecols=['gene','length'])
```

### Supplementary Fig. 6E: hybrid vs. short-read gene length distributions

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_plot = pd.DataFrame({
    'length': pd.concat([df_hy_g['length'], df_sr_g['length']], ignore_index=True),
    'assembly_type': ['long-read'] * len(df_hy_g) + ['short-read'] * len(df_sr_g)
})

fig, ax = plt.subplots(figsize=(3.5, 2.5), dpi=200)

sns.boxplot(data=df_plot, x='assembly_type', y='length')

plt.yscale('log')
plt.xlabel("")
plt.ylabel("gene length", fontweight='bold')
plt.title("")

plt.tight_layout()
plt.show()
```
