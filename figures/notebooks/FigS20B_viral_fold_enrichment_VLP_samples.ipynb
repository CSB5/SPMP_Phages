```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/bulk_vs_VLP_viral_fold_enrichment.tsv', sep='\t')

df.viral_enrichment.describe()
```

### Supplementary Fig. 20B: Viral fold enrichment in VLP samples
The viral fold enrichment in VLP samples is estimated as the ratio of Illumina read mapping rates to the SILVA release 138.1 SSU Ref NR99 database (at ≥95% identity over ≥80% of the read length) between the bulk vs. viral sample.

```python
df['log2_fold_enrichment'] = np.log2(df.viral_enrichment)
```

```python
sns.set_theme(style="whitegrid")

fig, ax = plt.subplots(figsize=(1.5, 3), dpi=200)

g = sns.violinplot(data=df, y='log2_fold_enrichment', cut=0)

plt.axhline(y=0, color='red')

plt.ylabel('viral fold enrichment', fontweight='bold')
plt.yticks(range(6),[1,2,4,8,16,32])

sns.despine(left=True, bottom=True)
```
