### Supplementary Fig. 5: agreement of VFCs with reference families

```python
import pandas as pd

df_v = pd.read_excel('../data/Supplementary_Data_1_vOTU_metadata.xlsx', usecols=['votu_id','VFC','host_phylum'])
df_vr = pd.read_excel('../data/Supplementary_Data_2_hierarchical_clustering_of_SPMP_and_reference_vOTUs.xlsx', usecols=['votu_id','VFC','dataset','Family','Subfamily'])
df_vc = pd.read_excel('../data/Supplementary_Data_3_VFC_metadata.xlsx', usecols=['VFC','VC_size'])
```

```python
def show_counts_proportions(fam, groups=['VFC']):
    return pd.concat(
        [
            fam[groups].value_counts(),
            fam[groups].value_counts(normalize=True)
        ],
        axis=1,
        keys=['count', 'proportion']
    )
```

"All best hits of ICTV-ratified families (underlined) with non-singleton VFCs are shown."

*Inoviridae* and *Winoviridae* are not tabulated as their best-hits are not a non-singleton VFC.

```python
nonsingletonVFCs = df_vc[df_vc.VC_size > 1].VFC.tolist()

df_vr[(df_vr.dataset == 'ICTV') & df_vr.VFC.isin(nonsingletonVFCs)][['Family','VFC']].value_counts(sort=False)
```
